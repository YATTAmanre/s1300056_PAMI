## Python Program
## Read csv file and data cleaning


```python
import pandas as pd
"""
A program read csv file and data cleaning

Function plus:
If value is less than or equal 0, replace this 0

Parameters
----------
plot_csv : dataframe
  Store Exercise9.csv as dataframe

dataset : dataframe
  Store Exercise9.csv data which clean data from Exercise9.csv
"""
plot_csv = pd.read_csv("./Exercise9.csv")

plot_csv.drop('Timestamp', inplace=True, axis=1)

sensorsWithPointInformation = [col for col in plot_csv if not col.startswith("POINT(")]
plot_csv.drop(columns=sensorsWithPointInformation, inplace=True, axis=1)

plot_csv.rename(columns=lambda x: x.replace("POINT(", "Point("), inplace=True)

for col in plot_csv.columns:
  plot_csv.loc[plot_csv[col] >= 100, col] = 0
  plot_csv.loc[plot_csv[col] <=0, col] = 0

dataset = plot_csv.fillna(0)

dataset
```

## Python program 4
## Implement FP-growth algorithm

```python
from PAMI.frequentPattern.basic import FPGrowth as alg
"""
inputFile : str
  Filename in input

minSup : int
  Minsup value in this FP-growth algorithm
"""

inputFile = 'PM24HeavyPollutionRecordingSensors.csv'

minSup=100

obj = alg.FPGrowth(inputFile, minSup)

obj.mine()

print("Total number of Frequent Patterns:", len(obj.getPatterns()))

obj.save('frequentPatterns.txt')

print("Total Memory in RSS", obj.getMemoryRSS())

print("Total ExecutionTime in seconds:", obj.getRuntime())
```
